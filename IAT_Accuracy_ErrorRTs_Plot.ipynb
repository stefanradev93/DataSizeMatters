```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import scipy.stats
```

```python
# get data
df = pd.read_csv("d:/iat/final_iat.csv")
```

```python
# aggregate age groups

df_grouped = df.groupby(['age']).agg(['mean','sem'])

df_grouped = df_grouped.iloc[:, [126,127,98,99,100,101]]
df_grouped.columns =df_grouped.columns.to_flat_index().map(lambda x: '_'.join(x))
df_grouped.reset_index(level=['age'], inplace=True)

df_grouped['49_mean'] = df_grouped['49_mean']*-1
df_grouped['50_mean'] = df_grouped['50_mean']*-1
```

```python
# Plot error RTs in congruent condition vs. age
f, ax= plt.subplots(figsize=(16, 12))

ax.scatter(df_grouped['age'], df_grouped['49_mean'], color='black', marker='o', alpha=.8)
ax.errorbar(df_grouped['age'], df_grouped['49_mean'], df_grouped['49_sem'], capsize=4, fmt='none',
                    color='black', alpha=0.3, linewidth=1)

sns.despine(ax=ax)
ax.set_ylabel('Error RT Congruent', fontsize=16)
ax.set_xlabel('Age', fontsize=16)
ax.set_ylim((0,4))
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
ax.grid(alpha=0.15)
f.tight_layout()
    
f.savefig('error_congruent.png', dpi=300)
```

```python
# Plot error RTs in incongruent condition vs. age

f, ax= plt.subplots(figsize=(16, 12))

ax.scatter(df_grouped['age'], df_grouped['50_mean'], color='black', marker='o', alpha=.8)
ax.errorbar(df_grouped['age'], df_grouped['50_mean'], df_grouped['50_sem'], capsize=4, fmt='none',
                    color='black', alpha=0.3, linewidth=1)

sns.despine(ax=ax)
ax.set_ylabel('Error RT Incongruent', fontsize=16)
ax.set_xlabel('Age', fontsize=16)
ax.set_ylim((0,4))
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
ax.grid(alpha=0.15)
f.tight_layout()
    
f.savefig('error_incongruent.png', dpi=300)
```

```python
# Plot accuracy vs. age

f, ax= plt.subplots(figsize=(16, 12))

ax.scatter(df_grouped['age'], df_grouped['Accuracy_mean'], color='black', marker='o', alpha=.8)
ax.errorbar(df_grouped['age'], df_grouped['Accuracy_mean'], df_grouped['Accuracy_sem'], capsize=4, fmt='none',
                    color='black', alpha=0.3, linewidth=1)

sns.despine(ax=ax)
ax.set_ylabel('Accuracy', fontsize=16)
ax.set_xlabel('Age', fontsize=16)
ax.set_ylim((.85,1))
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)
ax.grid(alpha=0.15)
f.tight_layout()
    
f.savefig('accuracy.png', dpi=300)
```
